```python
# to convert the weight into one unit

# to convert the weight from pound to kg
def convert_pounds_to_kg(weight):
    return float(weight.split()[0]) * 0.45359237

#to convert the weight from ounces to kg
def convert_ounces_to_kg(weight):
    return float(weight.split()[0]) * 0.0283495

def convert_to_kg(weight):
    if pd.isna(weight):
        return None
    elif 'pounds' in weight:
        return convert_pounds_to_kg(weight)
    elif 'ounces' in weight:
        return convert_ounces_to_kg(weight)
    else:
        return None
    
data['weight'] = data['weight'].apply(convert_to_kg)
```

```python
#find the mean of the weight where weight value is not zero

wt_mean = data['weight'][data['weight'] != 0.0].astype(float).mean()
wt_mean
```

```python
#import the required libraries for recommendation system algorithm
# we are going to use 

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#we are going to combine the below 3 columns as combined_feature and going to use to recomned the books 

data['combined_features'] = data['title'] + ' ' + data['author'] + ' ' + data['publisher'] 
```

```python
#to ceate the instance of 'TfidVectorizer' class  

tfidf_vectorizer = TfidfVectorizer()
```

```python
# Overall, this below code transform the textual data from combined_feature column of the dataframe into a TF_IDF matrix, 
# which can be used as input to machine learning agorithm for recommendation.

tfidf_matrix = tfidf_vectorizer.fit_transform(data['combined_features'])
```

```python
# the code computes the pairwise cosine similarity between all pairs of documents represented by their TF_IDF vectors, resulting
# in a symmetric similarity matrix where each element represents the cosine similarity between two documents, 
# which can be used for recommendation

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_book_recommend(book_index, similarity_matrix, data, top_n = 5):
    sim_scores = list(enumerate(similarity_matrix[book_index]))  # to calculate similarity score and save the list of tuple
    
    # This line sortd the list of similarity scores in descending order based on the similarity scores. 
    sim_scores = sorted(sim_scores, key = lambda x: x[1], reverse =True)
    
    sim_indices = [i[0] for i in sim_scores[1:top_n + 1]]  # to extract the book index and to store it
    
    recommended_books = data.iloc[sim_indices][['title', 'author', 'avg_reviews']]
    return recommended_books

book_title = 'Learning Tableau 2022: Create effective data visualizations, build interactive visual analytics, and improve your data storytelling capabilities, 5th Edition'
book_index = data[data['title'] == book_title].index[0]

recommendations = get_book_recommend(book_index, cosine_sim, data)
print(recommendations)
```

```python
# Create user-item interaction matrix
interaction_matrix = data[['title', 'avg_reviews', 'n_reviews', 'star5', 'star4', 'star3', 'star2', 'star1']]
interaction_matrix.fillna(0, inplace=True)

# Compute item-item similarity matrix (cosine similarity)
item_similarity = cosine_similarity(interaction_matrix.drop('title', axis=1))

# Get top-N similar items for each item
top_n = 5
similar_items = {}
for i in range(item_similarity.shape[0]):
    similar_indices = item_similarity[i].argsort()[-top_n-1:-1][::-1]
    similar_items[data['title'].iloc[i]] = data['title'].iloc[similar_indices].tolist()

# Recommend items to users based on similar items
def recommend_items(title, top_n=5):
    if title not in similar_items:
        return []
    return similar_items[title][:top_n]

# Example: Recommend items similar to 'Data Analysis Using R (Low Priced Edition)'
title = 'The Tableau Workshop: A practical guide to the art of data visualization with Tableau'
recommended_items = recommend_items(title, top_n=5)
print("Recommended Items for {}: {}\n".format(title, recommended_items))
```
